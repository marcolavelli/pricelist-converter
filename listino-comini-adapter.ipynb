```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('listino-comini.csv', sep=';')
```

```python
df.columns
```

```python
df = df[['Codice', 'Descrizione', 'Cod. Fornitore', 'U.M.', 'Listino', 'SC1', 'SC2', 'SC3', 'SC4', 'Netto']]
```

```python
df = df.fillna('0')
```

```python
df['Codice'] = df['Codice'].str.replace('Cod. ','')
```

```python
df['Codice'] = df['Codice'].str.replace('. ','')
```

```python
df['Cod. Fornitore'] = df['Cod. Fornitore'].str.replace('Cod. ','')
```

```python
df.to_csv('listino-comini.txt', header=None, index=None)
```
